### Running Sum of 1d Array

```python
# LeetCode Question link:
# https://leetcode.com/problems/running-sum-of-1d-array/

class Solution:
    def runningSum(self, nums: list[int]) -> list[int]:
        result = []
        temp = 0

        for i in nums:
            temp += i
            result.append(temp)

        return result
```

```python
obj = Solution()
print(obj.runningSum([2,9,6,7,5]))
```
