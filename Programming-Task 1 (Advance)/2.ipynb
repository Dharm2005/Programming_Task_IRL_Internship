### Shuffle the Array

```python
# LeetCode Question link:
# https://leetcode.com/problems/shuffle-the-array/

class Solution:
    def shuffle(self, nums: list[int], n: int) -> list[int]:
        result = []
        
        for i in range(n):
            result.append(nums[i])
            result.append(nums[n+i])

        return result
```

```python
obj = Solution()
print(obj.shuffle([1,5,3,4,8,7],3))
```
