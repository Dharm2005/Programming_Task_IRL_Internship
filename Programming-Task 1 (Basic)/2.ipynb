### Python If-Else

```python
if __name__ == '__main__':
    n = int(input().strip())  # My input is 15

    if not 1 <= n <= 100:
        print("number should be between 1 to 100")
    else:
        if n % 2 != 0:
            print("Weird")
        else:
            if 2 <= n <= 5:
                print("Not Weird")
            elif 6 <= n <= 20:
                print("Weird")
            else:
                print("Not Weird")
            
```
