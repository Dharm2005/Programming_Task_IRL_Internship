### Find Numbers with Even Number of Digits

```python
class Solution:
    def findNumbers(self, nums: list[int]) -> int:
        res = 0
        
        for i in nums:
          if len(str(i)) % 2 == 0:
            res += 1
            
        return res
```

```python
s = Solution()

print(s.findNumbers([121,3445,45,678,87]))
```
