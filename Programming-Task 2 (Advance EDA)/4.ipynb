### How Many Numbers Are Smaller Than the Current Number

```python
class Solution:
    def smallerNumbersThanCurrent(self, nums: list[int]) -> list[int]:
      res = []
      
      for i in nums:
        count = 0
        for j in nums:
          
          if j < i:
            count += 1
            
        res.append(count)
      
      return res
```

```python
s = Solution()

print(s.smallerNumbersThanCurrent([3,1,2,5,0]))
```
